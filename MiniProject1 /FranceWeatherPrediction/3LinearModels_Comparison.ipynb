#Dividing French Cities

```python
French_cities = ['MONTELIMAR', 'PERPIGNAN', 'TOURS']
French_columns = ['DATE', 'MONTH']

for city in French_cities:
    for col in df.columns:
        if col.startswith(city):
            French_columns.append(col)

df_French = df[French_columns]
print(df_French)
```

#Preprocessing

```python
date = df_French['DATE']
month=df_French['MONTH']
import pandas as pd
normalized_columns = df_French.columns.drop(['DATE', 'MONTH'])

for col in normalized_columns:
    min_val = df_French[col].min()
    max_val = df_French[col].max()

    df_French[col] = (df_French[col] - min_val) / (max_val - min_val)
df_French['DATE'] =date
df_French['MONTH']=month

print(df_French.head())
print(df_French.shape)
```

#Dividing Data from 2009

```python
df_French['DATE'] = pd.to_datetime(df_French['DATE'], format='%Y%m%d')
df_French= df_French[df_French['DATE'].dt.year == 2009]
print(df_French.head())
print(df_French.shape)
```

#Linear Models

##Target:Temprature

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
import numpy as np
import matplotlib.pyplot as plt


target_city = 'TOURS'
target_variable = 'temp_mean'

window_size = 5
overlap = 1
step_size = window_size - overlap
if step_size < 1:
    step_size = 1

def create_sliding_window_features(df, target_city, other_cities, target_variable, window_size, overlap):
    X = []
    y = []
    num_samples = len(df)
    step = window_size - overlap
    if step < 1:
        step = 1
    feature_names = []
    for i in range(0, num_samples - window_size, step):
        window = df.iloc[i:i + window_size]
        features = []
        current_feature_names = []
        for city in [target_city] + other_cities:
            if f'{city}_humidity' in df.columns:
                for t in range(window_size):
                    features.append(window[f'{city}_humidity'].iloc[t])
                    current_feature_names.append(f'{city}_humidity_t-{window_size - 1 - t}')
            if f'{city}_temp_mean' in df.columns:
                for t in range(window_size):
                    features.append(window[f'{city}_temp_mean'].iloc[t])
                    current_feature_names.append(f'{city}_temp_mean_t-{window_size - 1 - t}')
        X.append(features)
        if i + window_size < num_samples:
            y.append(df[f'{target_city}_{target_variable}'].iloc[i + window_size])
        if not feature_names and current_feature_names:
            feature_names = current_feature_names
    return np.array(X), np.array(y), feature_names

other_cities_sliding = [col.split('_')[0] for col in df_French.columns if col.startswith('MONTELIMAR') or col.startswith('PERPIGNAN')]
other_cities_sliding = list(set(other_cities_sliding))

X_sliding, y_sliding, sliding_feature_names = create_sliding_window_features(
    df_French, target_city, other_cities_sliding, target_variable, window_size, overlap
)

X_train_sliding, X_test_sliding, y_train_sliding, y_test_sliding = train_test_split(X_sliding, y_sliding, test_size=0.2, random_state=42, shuffle=False)


print("\n--- Ordinary Least Squares (OLS) with Sliding Window ---")
ols_model_sw = LinearRegression()
ols_model_sw.fit(X_train_sliding, y_train_sliding)
y_pred_ols_sw = ols_model_sw.predict(X_test_sliding)
mse_ols_sw = mean_squared_error(y_test_sliding, y_pred_ols_sw)
print(f"Mean Squared Error (OLS): {mse_ols_sw:.4f}")
print("OLS Coefficients:", ols_model_sw.coef_)
print("OLS Intercept:", ols_model_sw.intercept_)

print("\n--- Ridge Regression (L2 Regularization) with Sliding Window ---")
alpha_ridge = 1.0
ridge_model_sw = Ridge(alpha=alpha_ridge)
ridge_model_sw.fit(X_train_sliding, y_train_sliding)
y_pred_ridge_sw = ridge_model_sw.predict(X_test_sliding)
mse_ridge_sw = mean_squared_error(y_test_sliding, y_pred_ridge_sw)
print(f"Mean Squared Error (Ridge, alpha={alpha_ridge}): {mse_ridge_sw:.4f}")
print("Ridge Coefficients:", ridge_model_sw.coef_)
print("Ridge Intercept:", ridge_model_sw.intercept_)

print("\n--- Lasso Regression (L1 Regularization) with Sliding Window ---")
alpha_lasso = 0.1
lasso_model_sw = Lasso(alpha=alpha_lasso)
lasso_model_sw.fit(X_train_sliding, y_train_sliding)
y_pred_lasso_sw = lasso_model_sw.predict(X_test_sliding)
mse_lasso_sw = mean_squared_error(y_test_sliding, y_pred_lasso_sw)
print(f"Mean Squared Error (Lasso, alpha={alpha_lasso}): {mse_lasso_sw:.4f}")
print("Lasso Coefficients:", lasso_model_sw.coef_)
print("Lasso Intercept:", lasso_model_sw.intercept_)


plt.figure(figsize=(12, 6))
plt.plot(y_test_sliding, label='Actual Temperature', linewidth=2)
plt.plot(y_pred_ols_sw, label='OLS Prediction', linestyle='--')
plt.plot(y_pred_ridge_sw, label=f'Ridge Prediction (alpha={alpha_ridge})', linestyle='-.')
plt.plot(y_pred_lasso_sw, label=f'Lasso Prediction (alpha={alpha_lasso})', linestyle=':')
plt.xlabel('Time Steps')
plt.ylabel('Temperature')
plt.title(f'Temperature Prediction for {target_city} using Different Linear Regression Methods (Sliding Window)')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(14, 7))
num_coeffs = len(sliding_feature_names)
bar_width = 0.2

index = np.arange(num_coeffs)

plt.bar(index - bar_width, ols_model_sw.coef_, bar_width, label='OLS Coefficients')
plt.bar(index, ridge_model_sw.coef_, bar_width, label=f'Ridge Coeff. (alpha={alpha_ridge})')
plt.bar(index + bar_width, lasso_model_sw.coef_, bar_width, label=f'Lasso Coeff. (alpha={alpha_lasso})')

plt.xlabel('Lagged Features')
plt.ylabel('Coefficient Value')
plt.title('Comparison of Coefficients for Different Regression Methods (Sliding Window)')
plt.xticks(index, sliding_feature_names, rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```
